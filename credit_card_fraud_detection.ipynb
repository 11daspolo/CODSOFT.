```python
#import some libreries
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#import dataset
credit_card_dataset = pd.read_csv('/content/sample_data/creditcard.csv')
credit_card_dataset.head()
```

```python
credit_card_dataset.sample()
```

```python
#credit card data information
credit_card_dataset.info()
```

```python
# checking the number of missing values in each column
credit_card_dataset.isnull().sum()
```

```python
# distribution of legit transactions & fraudulent transactions
credit_card_dataset['Class'].value_counts()
```

```python

fraud = credit_card_dataset[credit_card_dataset['Class']==1]
legit = credit_card_dataset[credit_card_dataset.Class==0]
```

```python
fraud['Class']
```

```python
#statistical measurement
fraud.Amount.describe()
```

```python
# statistical measurement
legit.Amount.describe()
```

```python

credit_card_dataset.groupby('Class').mean()
```

```python
legit_sample = legit.sample(n=395)
new_df = pd.concat([legit_sample,fraud],axis=0)
new_df
```

```python
new_df['Class'].value_counts()
```

```python
new_df.groupby('Class').mean()
```

```python
X = new_df.drop(columns='Class', axis=1)
Y = new_df['Class']
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=4)
```

```python
#logisticregression
model=LogisticRegression()
```

```python
# training the Logistic Regression Model
model.fit(X_train, Y_train)
```

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
print('Accuracy score on Test Data : ', test_data_accuracy)
```

```python

# accuracy on train data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
print('Accuracy on Train data : ', training_data_accuracy)
```
